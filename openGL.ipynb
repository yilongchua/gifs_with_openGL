## Defining a cube

8 nodes with 8 lines 

Assumptions:
1. lines are of equal length
2. there must be 8 sides
3. ** starting point must always be on bottom left front corner** this is marked in red
4. starting point must be define in terms of coordinates system, i.e (X,Y,Z) = (1,1,1)

# cube function

```python
def cube(start_point,length = 1 ):
#     print("location of starting point X:{}, Y:{}, Z:{}".format(start_point[0],start_point[1],start_point[2]))
#     print('This cube has a length of {}'.format(length))
    p0 = start_point
    p1 = (start_point[0] + length,) + start_point[1:]
    p2 = (start_point[0], start_point[1] + length, start_point[2])
    p3 = (start_point[0] + length, start_point[1] + length, start_point[2])
    p4 = start_point[:2] + (start_point[2] + length,)
    p5 = (start_point[0] + length,) + (start_point[1],) + (start_point[2] + length,)
    p6 = (start_point[0],) + (start_point[1] + length, start_point[2] + length)
    p7 = (start_point[0] + length, start_point[1] + length, start_point[2] + length)
    
#     print(p0,p1,p2,p3,p4,p5,p6,p7)
    vertices = (p0,p1,p2,p3,p4,p5,p6,p7)
    edges = ((0,1),(0,2),(0,4),(2,3),(1,3),(2,6),(3,7),(1,5),(4,5),(4,6),(6,7),(5,7))
    
#     for edge in edges:
#         print('edge',edge)
#         for vertex in edge:
#             print('vertices[vertex] = {}'.format(vertices[vertex]))
    
    glBegin(GL_LINES)
    for edge in edges:
        for vertex in edge:
            glVertex3fv(vertices[vertex])
    glEnd()

```

# main running function to form rotating cube

```python
def main():
    pygame.init()
    display = (800,600)
    pygame.display.set_mode(display, DOUBLEBUF|OPENGL)

    gluPerspective(45, (display[0]/display[1]), 0.1, 50.0)

    glTranslatef(0.0,0.0, -5)

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()

        glRotatef(1, 3, 1, 1)
        glClear(GL_COLOR_BUFFER_BIT|GL_DEPTH_BUFFER_BIT)
        cube(start,1)
        pygame.display.flip()
        pygame.time.wait(10)

```

## Creating gifs

```python
#Import Libraries
import pygame
import OpenGL
from pygame.locals import *
from OpenGL.GL import *
from OpenGL.GLU import *
import glob
from PIL import Image
import math
import copy
import os
from os import listdir
from os.path import isfile, join
from os import walk


# define some params for pygame screen
pygame.init()
display = (800,600)
screen = pygame.display.set_mode(display, DOUBLEBUF | OPENGL)
pygame.display.set_caption('Apex Centoid Triangles')

#define pyramids
# assume all sides to be the same length
def pyramid(start_point, length=1):
    #apex
    p0 = start_point
    #base
    p1 = (start_point[0] - length*0.5, start_point[1] - length, start_point[2] - length*0.5)
    p2 = (start_point[0] + length*0.5, start_point[1] - length, start_point[2] - length*0.5)
    p3 = (start_point[0] - length*0.5, start_point[1] - length, start_point[2] + length*0.5)
    p4 = (start_point[0] + length*0.5, start_point[1] - length, start_point[2] + length*0.5)
    
#     print(p0,p1,p2,p3,p4)
    vertices = (p0,p1,p2,p3,p4)
    edges = ((0,1),(0,2),(0,3),(0,4),(1,2),(1,3),(2,4),(3,4))
    
    glBegin(GL_LINES)
    for edge in edges:
        for vertex in edge:
            glVertex3fv(vertices[vertex])
    glEnd()
    
# assume all sides to be the same length
def inv_pyramid(start_point, length=1):
    #apex
    p0 = start_point
    #base
    p1 = (start_point[0] - length*0.5, start_point[1] + length, start_point[2] - length*0.5)
    p2 = (start_point[0] + length*0.5, start_point[1] + length, start_point[2] - length*0.5)
    p3 = (start_point[0] - length*0.5, start_point[1] + length, start_point[2] + length*0.5)
    p4 = (start_point[0] + length*0.5, start_point[1] + length, start_point[2] + length*0.5)
    
#     print(p0,p1,p2,p3,p4)
    vertices = (p0,p1,p2,p3,p4)
    edges = ((0,1),(0,2),(0,3),(0,4),(1,2),(1,3),(2,4),(3,4))
    
    glBegin(GL_LINES)
    for edge in edges:
        for vertex in edge:
            glVertex3fv(vertices[vertex])
    glEnd()
    
                        
def add_pyramids_x_z(number,direction,start1, start2):
        
    for i in range(int(number+1)):
        #north -z
        if direction.lower() =='n':
            start_dir1 = start1[:2]+ (start1[2] - i,)
            start_dir2 = start2[:2]+ (start2[2] - i,)
            pyramid(start_dir1,length)
            inv_pyramid(start_dir2,length)

        #south +z
        if direction.lower() =='s':
            start_dir1 = start1[:2]+ (start1[2] + i,)
            start_dir2 = start2[:2]+ (start2[2] + i,)
            pyramid(start_dir1,length)
            inv_pyramid(start_dir2,length)
         #east +x
        if direction.lower() =='e':
            start_dir1 = (start1[0]+ i,) + start1[1:]
            start_dir2 = (start2[0]+ i,) + start2[1:]
            pyramid(start_dir1,length)
            inv_pyramid(start_dir2,length)
        #west -x
        if direction.lower() =='w':
            start_dir1 = (start1[0]- i,) + start1[1:]
            start_dir2 = (start2[0]- i,) + start2[1:]
            pyramid(start_dir1,length)
            inv_pyramid(start_dir2,length)
            

def add_pyramids_layers_y(direction, current_loop,frame_interval = 300,):
    
    #length of 2 pyramids
    full_length = (math.sqrt(length**2 + (0.5*length)**2))*2 + 0.05
    
    #when to activate the next layer
    layer_no = loop_number(current_loop, frame_interval) + 1
    
    if layer_no:
        
        for i in range(1,layer_no):
            #up +y
            if direction.lower() =='up':
                start_dir1 = (start1[0], start1[1] + i*full_length, start1[2])
                start_dir2 = (start2[0], start2[1] + i*full_length, start2[2])
                compass_directions_pyramids(i,start_dir1,start_dir2)

                
            #down -y
            elif direction.lower() =='down':
                start_dir1 = (start1[0], start1[1] - i*full_length, start1[2])
                start_dir2 = (start2[0], start2[1] - i*full_length, start2[2])
                
                compass_directions_pyramids(i,start_dir1,start_dir2)
           
                            
                
def loop_number(loop, frame_interval = 300):
    return loop//frame_interval

def compass_directions_pyramids(layer_no, start_dir1, start_dir2):
    #middle pyramid
    pyramid(start_dir1,length)
    inv_pyramid(start_dir2,length)
    #4 direction adding pyramid
    ls_dir = ['n', 's', 'e', 'w']
    for direction in ls_dir:
        add_pyramids_x_z(layer_no,direction,start_dir1, start_dir2)
        
def save_image(screen, frame_interval,loop,filename):
    if loop % frame_interval == 0:
        pygame.image.save(screen, f"{filename}{int(loop/frame_interval)}.png")
        print(loop, 'image saved' )
       
        
        
def save_gifs(filename):
    frames = []
    #get list of images in folder
    ls = [int(f[:-4]) for f in listdir(filename) if '.png' in f]
    ls.sort()
    imgs = [(f'{filename}' + str(i)+'.png') for i in ls]
    for count,i in enumerate(imgs):        
        frame = Image.open(i)
        keep = frame.copy()
        frame.close()
        frames.append(keep)
        
    frames[0].save('layered pyramid.gif', format='GIF',
                                   append_images=frames[1:],
                                   save_all=True,
                                   duration=25, loop=0)
    print(len(frames),'images are saved into gif')
    

# define main function to run    
def main():
    gluPerspective(45, (display[0]/display[1]), 0.1, 50.0)

    glTranslatef(0,-10, -20)
    
    loop = 0
    frame_interval = 300
    
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
                       
        glTranslatef(0,0.05 + loop_number(loop, 300)/100, 0)
        glRotatef(1, 0, 1, 0)
        glClear(GL_COLOR_BUFFER_BIT|GL_DEPTH_BUFFER_BIT)
        
        #middle
        pyramid(start1,length)
        inv_pyramid(start2,length)
        
        add_pyramids_layers_y('down',loop,50)
#         add_pyramids_layers_y('up',loop, 300)
            
        save_image(screen,1,loop,filename)
    
        if event.type == pygame.KEYDOWN:    
            #save image
            if event.key == pygame.K_s:
                save_image(screen,2,loop,filename)
            #save gifs
            if event.key == pygame.K_g:
                save_gifs(filename)
        
        loop +=1
        pygame.display.flip()
        pygame.time.wait(10)
        
        

#middle
start1 = (0,1.05,0)
start2 = (0,-1.05,0)
length = 1
filename = './apex pyramid w layers/'
main()
```
